```python
import pandas as pd
import numpy as np

cols = ['col0', 'col1', 'col2', 'col3', 'col4', 'col5', 'col6']
rows = ['row0', 'row1', 'row2', 'row3', 'row4']
data = np.random.randint(0, 50, size=(5,7))
d = pd.DataFrame(data, columns=cols, index=rows)

d.head()
```

```python
# AGGREGATING DATA IN 2D ARRAY
```

```python
d = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['X', 'Y', 'Z'])
```

```python
d.agg(['sum', 'max'])
```

```python
d.agg({'X' : ['sum', 'min'], 'Y' : ['min', 'max']})
```
